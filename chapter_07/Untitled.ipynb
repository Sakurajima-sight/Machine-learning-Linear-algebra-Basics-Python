```python
from sympy import cos,sin
from sympy import integrate
from sympy.abc import x
from scipy.fftpack import fft
import numpy as np
import matplotlib.pyplot as plt
```

```python
e=integrate(sin(x)*cos(x),(x,0,2*np.pi))
print(e.evalf())
```

```python
e1=integrate(sin(2*x)*cos(5*x),(x,0,2*np.pi))
e2=integrate(sin(4*x)*cos(0*x),(x,0,2*np.pi))
e3=integrate(sin(x)*cos(2*x),(x,0,2*np.pi))
print(e1.evalf())
print(e2.evalf())
print(e3.evalf())
```

```python
z1=2+1j
z2=3+5j
# 实部
print(z1)
# 虚部
print(z2)
# 加法
print(z1+z2)
# 乘法
print(z1*z2)
# 共轭
print(z1.conjugate())
# 模长
print(abs(z1))
```

```python
def f(x):
    return np.sin(x)+2*np.sin(3*x)+2*np.cos(3*x)+4*np.sin(15*x)

x=np.linspace(0,2*np.pi,2048)
plt.scatter(x,f(x))
plt.grid()
plt.show()
```

```python
x=np.linspace(0,2*np.pi,128)
y=np.sin(x)+2*np.sin(3*x)+2*np.cos(3*x)+4*np.sin(15*x)

# 离散频率
xf=np.arange(len(y))
# 由于对称性 因此只取一半区域
xf_half=xf[range(int(len(x)/2))]
# 执行完fft后 对各频率的能量进行归一化处理
yf=abs(fft(y))/len(x)
# 由于对称性 因此只取一半区域
yf_half=yf[range(int(len(x)/2))]

plt.plot(xf_half,yf_half)
plt.show()
```
